```python
row1 = [278,240,284,147,281,226,342,349,162,326,307,329,262,301,307,323,96,349,296,262,165,259,277,250,245,276,234,310,125,291,258,286,261,119,240]
row2 = [666,512,650,377,627,484,796,815,422,760,719,757,556,709,715,767,224,809,694,556,431,579,603,548,541,596,500,730,311,635,564,626,575,293,526]

A = np.array([2, 1, 4, 3]).reshape(2, 2)
B = np.array([row1, row2])

A_inv = np.linalg.inv(A)

B_original = A_inv.dot(B)

B_original = np.rint(B_original).astype(int)


for i in B_original:
  for j in i:
    print(chr(j), end="")
```

# Cipher 4 - RSA
---

```python
def factor(n):
    # try dividing by all numbers up to sqrt(n)
    limit = int(n**0.5) + 1
    for i in range(2, limit):
        if n % i == 0:
            return i, n // i
    return None
```

```python
def modinverse(e: int, phi: int) -> int:
  e = e % phi
  for x in range(1, phi):
    if ((e * x) % phi == 1):
      return x
  return 1
```
